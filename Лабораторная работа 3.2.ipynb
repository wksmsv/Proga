```python
# Задание task_03_02_02.
#
# Выполнил: Студент
# Группа: !!!


def foo(i):
    """Алгоритм: преобразование целого неотрицательного числа в строку
    (аналог встроенной функции str) путём последовательного выделения
    последней цифры и добавления её в начало результирующей строки.

    Параметры:
        - i (int): целое неотрицательное число.

    Сложность: O(log N), где N — значение числа i (количество итераций
    равно количеству цифр в числе, т.е. ⌊log10(N)⌋+1).
    """
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result
```
