```python
# Задание task_03_04_02 и task_03_04_01
#
# Выполнил: Самусева В.А.
# Группа: ЦИБ-251

"""

3.4.1: Ввод вещественных чисел с клавиатуры, запись каждого числа на отдельной строке в файл.
3.4.2: Чтение файла, вычисление суммы и максимума, дозапись результатов в конец файла.
       При ошибке чтения – сообщить и завершить программу.
"""

def write_numbers_to_file(filename: str) -> None:
    """Задача 3.4.1: запрос чисел и запись в файл."""
    user_input = input("Введите вещественные числа через пробел: ")
    # Разбиваем строку на части и преобразуем в float
    try:
        numbers = [float(x) for x in user_input.split()]
    except ValueError:
        print("Ошибка: введены некорректные числа. Завершение программы.")
        return

    # Запись в файл (режим 'w' – перезапись)
    with open(filename, 'w', encoding='utf-8') as f:
        for num in numbers:
            f.write(f"{num}\n")
    print(f"Числа записаны в файл '{filename}'.")


def process_file(filename: str) -> None:
    """Задача 3.4.2: чтение чисел из файла, вычисление суммы и максимума, дозапись."""
    # 1. Попытка чтения файла
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            lines = f.readlines()
    except (FileNotFoundError, IOError) as e:
        print(f"Ошибка при чтении файла '{filename}': {e}")
        return  # Прекращаем выполнение, как требуется в условии

    # 2. Извлечение вещественных чисел из строк
    numbers = []
    for line in lines:
        line = line.strip()
        if line:   # игнорируем пустые строки
            try:
                numbers.append(float(line))
            except ValueError:
                # По условию 3.4.2 файл должен содержать только числа,
                # но если встретится что-то иное – пропускаем с предупреждением
                print(f"Предупреждение: пропущена некорректная строка: '{line}'")

    if not numbers:
        print("Файл не содержит чисел. Обработка невозможна.")
        return

    # 3. Вычисления
    total = sum(numbers)
    maximum = max(numbers)

    # 4. Дозапись результатов в файл
    try:
        with open(filename, 'a', encoding='utf-8') as f:
            f.write(f"{total}\n{maximum}\n")
        print(f"Результаты добавлены в '{filename}':")
        print(f"  Сумма = {total}")
        print(f"  Максимум = {maximum}")
    except IOError as e:
        print(f"Ошибка при записи в файл: {e}")


def main():
    filename = "numbers.txt"

    # Выполняем задачу 3.4.1
    write_numbers_to_file(filename)

    # Выполняем задачу 3.4.2 (обрабатываем только что созданный файл)
    process_file(filename)


if __name__ == "__main__":
    main()
```
