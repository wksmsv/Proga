```python
# Программирование на языке высокого уровня (Python).
# Задание task_02_22.
#
# Выполнил: Самусева В.А.
# Группа: ЦИБ-251
# E-mail: samusevava489@mgpu.ru

#Даны n вещественных чисел. Определите максимальное и минимальное из них. 23При выводе вещественных результатов оставьте два знака после запятой.


n = int(input("Введите количество чисел: "))

a_max = None
a_min = None

for i in range(n):
    x = float(input(f"Введите {i+1}-е число: "))
    if i == 0:
        a_max = a_min = x
    else:
        if x > a_max:
            a_max = x
        if x < a_min:
            a_min = x

print("Максимум: {:.2f}".format(a_max))
print("Минимум: {:.2f}".format(a_min))

# --------------
# Пример вывода:
#
# n = 4
# 1-е число = 6.2
# 2-е число = 3.8
# 3-е число = 1.1
# 4-е число = 9.66
# Максимум: 9.66
```
